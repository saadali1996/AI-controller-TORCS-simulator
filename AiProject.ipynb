```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
import _pickle as cPickle
from keras.models import load_model
from sklearn.externals import joblib
```

```python
xframe=pandas.read_csv("X.csv", header=None)
```

```python
xframe.describe()
```

```python
yframe=pandas.read_csv("Y.csv", header=None)
```

```python
yframe.describe()
```

```python
dt1=xframe.values
dt2=yframe.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(dt1, dt2, test_size=0.1, random_state=42)
```

```python
max_depth=30
```

```python
regr_multirf = RandomForestRegressor(n_estimators=10,max_depth=max_depth,max_features='sqrt',
                                                          random_state=0,n_jobs=-1,verbose=False)
```

```python
regr_multirf.fit(X_train, y_train)
```

```python
regr_multirf.score(X_test,y_test)
```

```python
with open('MultiOutputRegressor4.pkl', 'wb') as fid:
    cPickle.dump(regr_multirf, fid) 
```

```python
from sklearn.ensemble import GradientBoostingRegressor
cl2=MultiOutputRegressor(GradientBoostingRegressor(verbose=2), n_jobs=-1).fit(X_train, y_train)
```

```python
print(cl2.score(X_test,y_test))
```

```python
model = Sequential()
model.add(Dense(64, input_dim=79, kernel_initializer='normal', activation='sigmoid'))
model.add(Dense(256, kernel_initializer='normal', activation='relu'))
model.add(Dense(128, kernel_initializer='normal', activation='relu'))
model.add(Dense(7, kernel_initializer='normal'))
# Compile model
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train,          
          nb_epoch=200,batch_size=2000, validation_split=0.2, verbose = 1)
```

```python
score = model.evaluate(X_test, y_test, verbose=0) 
```

```python
print(score)
```
