```python
dt1=merged.values
dt2=merged1.values
X_train, X_test, y_train, y_test = train_test_split(dt1, dt2, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Dense(128, input_dim=79, kernel_initializer='normal'))
model.add(Activation('relu'))
model.add(Dense(64, kernel_initializer='normal', activation='relu'))
model.add(Dense(128, kernel_initializer='normal', activation='relu'))
model.add(Dense(64, kernel_initializer='normal', activation='relu'))
model.add(Dense(7, kernel_initializer='normal'))
# Compile model
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['accuracy'])
```

```python
tbCallBack = keras.callbacks.TensorBoard(log_dir='/tmp/keras_logs', write_graph=True)
```

```python
model.fit(X_train, y_train,          
          nb_epoch=200,batch_size=2000, validation_split=0.2, verbose = 1,callbacks=[tbCallBack])
```

```python
score = model.evaluate(X_test, y_test, verbose=0) 
print(score)
```
